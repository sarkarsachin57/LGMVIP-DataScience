#### AIM : Converting an Image to its pencil scketch version.

>

#### Importing Packages

```python
import cv2
import matplotlib.pyplot as plt
```

#### Reading Image

```python
img = cv2.imread('kolkata.jpg')
```

#### Real Image - Kolkata

```python
plt.figure(figsize=(15,10))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB),cmap='gray')
plt.show()
```

#### Image to Grayscale version

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(15,10))
plt.imshow(gray,cmap='gray')
plt.show()
```

#### Inverted version

```python
inverted = 255 - gray
plt.figure(figsize=(15,10))
plt.imshow(inverted,cmap='gray')
plt.show()
```

#### Blur Inverted Version

```python
blur = cv2.GaussianBlur(inverted,ksize=(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(15,10))
plt.imshow(blur,cmap='gray')
plt.show()
```

#### Pencil Scketch Version

```python
pencil = cv2.divide(gray,255-blur,scale=256)
plt.figure(figsize=(15,10))
plt.imshow(pencil,cmap='gray')
plt.show()
```
